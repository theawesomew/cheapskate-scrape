```python
import selenium
import selenium.webdriver
import selenium.webdriver.common
import selenium.webdriver.common.by
import selenium.webdriver.support
import selenium.webdriver.support.expected_conditions
import selenium.webdriver.support.wait
import re

options = selenium.webdriver.ChromeOptions()
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:133.0) Gecko/20100101 Firefox/133.0')
webdriver = selenium.webdriver.Chrome(options=options)
wait = selenium.webdriver.support.wait.WebDriverWait(webdriver, 10)

try:
    for i in range(1, 15):
        webdriver.get(f'https://www.woolworths.com.au/shop/browse/fruit-veg?pageNumber={i}')
        webdriver.execute_script('window.scrollTo(0, window.innerHeight)')
        els = wait.until(selenium.webdriver.support.expected_conditions.presence_of_all_elements_located((selenium.webdriver.common.by.By.TAG_NAME, 'wc-product-tile')))

        with open(f'fruit-veg-{i}.csv', 'w+') as f:
            for el in els:
                l = list(reversed(el.text.split('\n')[1:-2]))

                if len(l) != 2:
                    continue

                f.write(f'{l[0]}, {l[1]}\n')
finally:
    webdriver.quit()
```
